```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
from splinter import Browser
import time
import json
import re
```

```python
# Defining functions to instantiate webdriver, opening browser and closing browser.

def init_browser():
    # Import Splinter and set the chromedriver path
    executable_path = {"executable_path": "chromedriver.exe"}
    browser = Browser("chrome", **executable_path, headless=False)
    return browser

def open_browser(browser, url):
    browser.visit(url)

def close_browser(browser):
    browser.quit()
```

```python
url = "https://www.thedailymeal.com/eat/best-pizza-every-state-slideshow"
```

```python
# Use splinter to get the html page into an object
browser = init_browser()
open_browser(browser, url)
```

```python
# Use bs4 to scrape the elemnets from the html object
state = []
name = []
city = []
# <div class="image-title slide-title"><h2>Alabama: Bottega Café (Birmingham)</h2></div>
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
results = soup.find_all('div', {'class':'slide-main'})
print(len(results))

for item in results:
    state_string = item.find('div', {'class' : ['image-title', 'slide-title']}).find('h2').text
    try:
        part_one = state_string.split(':')[0]
        part_two = state_string.split(':')[1]
        pattern = re.compile('(|)')
        if re.search(pattern,part_two).group(0):
            #print(state_string)
            part_three = part_two.split('(')[1].split(')')[0]
            part_four =  part_two.split('(')[0]
        else:
            #print(state_string)
            part_three = part_two
            part_four = part_one
        state.append(part_one)
        city.append(part_three)
        name.append(part_four)
    except Exception as e:
        print(f"Skipping....")


        
    print(f"{part_one} --{part_two}- {part_three} -- {part_four}")


    
    
```

```python
state_string = "Washington, D.C.: 2Amys "
part_one = state_string.split(':')[0]
part_two = state_string.split(':')[1]
#print( part_one,  part_two)
print(re.search(r'(|)',part_two).group(0))
```

```python
# Creating a dataframe for state wise best pizzas
best_pizza_stores = pd.DataFrame({'best_resaurant' : name,
                                 'city' : city,
                                'state': state })
best_pizza_stores
```
